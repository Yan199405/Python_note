### 基本使用

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.wait import WebDriverWait

browser = webdriver.Chrome()
try:
    browser.get('http://www.baidu.com')
    input = browser.find_element_by_id('kw')
    input.send_keys('python')
    input.send_keys(Keys.ENTER)
    wait = WebDriverWait(browser,10)
    wait.until(EC.presence_of_element_located((By.ID,'content_left')))
    print(browser.current_url)
    print(browser.get_cookies())
    print(browser.page_source)
finally:
    browser.close()
    
```

### 查找元素
#### 单个元素

```python
from selenium import webdriver
browser = webdriver.Chrome()
browser.get('https://www.baidu.com')
input1 = browser.find_element_by_id('s_tab')
input2 = browser.find_element_by_css_selector('#s_tab')
input3 = browser.find_element_by_xpath('//*[@id="s_tab"]')    #这三种选择器选定的内容是一样的
print(input1,'\n',input2,'\n',input3)
browser.close()
```

- find_element_by_name
- find_element_by_xpath
- find_element_by_link_text
- find_element_by_partial_link_text
- find_element_by_tag_name
- find_element_by_class_name
- find_element_by_css_selecter

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
browser = webdriver.Chrome()
browser.get('https://www.baidu.com')
input = browser.find_element(By.ID,'s_tab')     #这种方法结果等同于input1 = browser.find_element_by_id('s_tab')
print(input)
browser.close()
```

### 元素的交互操作
- https://selenium-python.readthedocs.io/api.html#module-selenium.webdriver.remote.webelement

```python
#案例：自动操作淘宝
from selenium import webdriver
import time

browser = webdriver.Chrome()
browser.get('http://www.taobao.com')
input = browser.find_element_by_id('q')   
input.send_keys('iphone')  #输入框输入iPhone
time.sleep(1) 
input.clear()           #清除输入
input.send_keys('iPad')  #输入iPad
button = browser.find_element_by_class_name('btn-search')   #查找button按钮元素
button.click()   #模仿鼠标左击，搜索
```

#### 交互动作
- 将动作附加到动作链中串行执行
- https://selenium-python.readthedocs.io/api.html#module-selenium.webdriver.common.action_chains

```python
#案例：创建一个动作链，执行拖拽动作
from selenium import webdriver
from selenium.webdriver import ActionChains
browser = webdriver.Chrome()
url = 'https://www.runoob.com/try/try.php?filename=jqueryui-api-droppable'
browser.get(url)
browser.switch_to.frame('iframeResult')
source = browser.find_element_by_id('draggable')
target = browser.find_element_by_id('droppable')
actions = ActionChains(browser)
actions.drag_and_drop(source,target)
actions.perform()
```

### Frame

```python
#案例：
import time
from selenium import webdriver
from selenium.webdriver import ActionChains
from selenium.common.exceptions import NoSuchElementException
browser = webdriver.Chrome()
url = 'https://www.runoob.com/try/try.php?filename=jqueryui-api-droppable'
browser.get(url)
browser.switch_to.frame('iframeResult')
source = browser.find_element_by_id('draggable')
print(source)
try:
    logo = browser.find_element_by_class_name('logo')
except NoSuchElementException:
    print('NO LOGO')
browser.switch_to.parent_frame()     ##获取父frame
logo = browser.find_element_by_class_name('logo')
print(logo)
print(logo.text)
```

### 等待
#### 隐式等待
- 当使用的隐式等待执行测试的时候，如果WebDriver没有在DOM中找到元素，将继续等待,超出设定时间后则抛出找不到元素的异常，换句话说，当查找元素或元素并没有立即出现的时候，隐式等待将等待一段时间再查找DOM，默认时间是0

```python
from selenium import webdriver

browser = webdriver.Chrome()
browser.implicitly_wait(10)    #如果元素加载出来则不等待，如果没有则等待10s
browser.get('http://www.zhihu.com/explore')
input = browser.find_element_by_class_name('zu-top-nav-link')
print(input)
```

#### 显示等待
- 规定一个等待条件，一个最长等待时间，如果超过这个最长时间条件还没有达到，则抛出异常

```python
from selenium import webdriver 
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
browser = webdriver.Chrome()
browser.get('http://www.taobao.com')
wait = WebDriverWait(browser,10)
input = wait.until(EC.presence_of_all_elements_located((By.ID,'q')))   ##如果获取到‘q’元素则正常，没有的话则等待
button = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR,'.btn-search')))    ###获取输入框和按钮
print(input,'\n',button)
browser.close()
```

### 选项卡管理

### 异常处理
- https://selenium-python.readthedocs.io/api.html#module-selenium.common.exceptions

```python
from selenium import webdriver
browser = webdriver.Chrome()
browser.get('http://www.baidu.com')
browser.find_element_by_id('hello')

```

```python
from selenium import webdriver
from selenium.common.exceptions import TimeoutException,NoSuchElementException

browser = webdriver.Chrome()
try:
    browser.get('http://www.baidu.com')
except TimeoutException:
    print('No Element')
try:
    browser.find_element_by_id('hello')
except NoSuchElementException:
    print('No Element')
finally:
    browser.close()
```
